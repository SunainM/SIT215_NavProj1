```python
from pdb import run
from typing_extensions import runtime
import pygame
from pytmx import load_pygame
from enum import Enum, auto
import heapq
import math
from typing import List, Tuple, Optional, Any, Dict
import time

# -----------------------------------------------------------------------------
# Main problem Class with all inner classes and helper functions
# -----------------------------------------------------------------------------
class Problem:
    """
    The main problem class for the wheelchair map navigation system.

    This class encapsulates all helper classes (Tile, Button, Level, Map, PriorityQueue,
    StateType, State, Agent, Analyzer) as inner classes. It also manages private configuration
    variables for Dijkstra's friction usage, agent speed, and the A* heuristic multiplier.
    Public setter and getter methods are provided for these settings. The problem class handles
    state transitions, rendering, and event handling through its main loop.
    """

    # -------------------------------------------------------------------------
    # Inner Helper Functions (Static Methods)
    # -------------------------------------------------------------------------
    @staticmethod
    def get_tile_center(tile: "Problem.Tile") -> Tuple[int, int]:
        """
        Calculate and return the center coordinates of the given tile.
        
        Args:
            tile (problem.Tile): The tile for which to compute the center.
            
        Returns:
            Tuple[int, int]: The (x, y) coordinates of the tile's center.
        """
        center_x = tile.pos_x * tile.width + tile.width // 2
        center_y = tile.pos_y * tile.height + tile.height // 2
        return center_x, center_y

    @staticmethod
    def calculate_cost(source: "Problem.Tile", neighbor: "Problem.Tile", use_terrain_cost: bool) -> float:
        """
        Calculate the movement cost from the source tile to a neighboring tile.

        Args:
            source (problem.Tile): The starting tile.
            neighbor (problem.Tile): The adjacent tile.
            use_terrain_cost (bool): Flag to determine if additional terrain factors should be used.

        Returns:
            float: The computed move cost.
        """
        elevation_diff = abs(neighbor.z_level - source.z_level)
        if use_terrain_cost:
            return (source.friction + neighbor.friction) / 2 + elevation_diff + 1
        else:
            return elevation_diff + 1

    # -------------------------------------------------------------------------
    # Inner Classes
    # -------------------------------------------------------------------------
    class Tile:
        """
        Represents a single tile on the map.
        
        Attributes:
            pos_x (int): The x index of the tile.
            pos_y (int): The y index of the tile.
            level (int): The map level of the tile.
            image: The pygame Surface used to represent the tile.
            properties (dict): Dictionary containing additional attributes (e.g., id, friction, zone).
            friction (float): Surface friction value.
            zone (str): A string indicating the tile's zone (used for obstacles or elevators/stairs).
            height (int): The height of the tile.
            z_level (float): The elevation level of the tile (for cross-level transitions).
            location: Additional information about the tileâ€™s location.
            width (int): The width of the tile.
        """
        def __init__(self, pos_x: int, pos_y: int, level: int, image: Any, properties: Dict[str, Any]) -> None:
            self.pos_x = pos_x
            self.pos_y = pos_y
            self.level = level
            self.image = image
            self.properties = properties
            self.gid = properties['id']
            self.friction = properties['SurfaceFriction']
            self.zone = properties['Zone']
            self.height = properties['height']
            self.z_level = float(level)
            self.location = properties['location']
            self.width = properties['width']

        def __eq__(self, other: Any) -> bool:
            return isinstance(other, Problem.Tile) and (self.pos_x, self.pos_y, self.level) == (other.pos_x, other.pos_y, other.level)

        def __hash__(self) -> int:
            return hash((self.pos_x, self.pos_y, self.level))

        def __lt__(self, other: "Problem.Tile") -> bool:
            # Tiles are compared first by elevation (z_level), then by y-coordinate, then x-coordinate.
            return (self.z_level, self.pos_y, self.pos_x) < (other.z_level, other.pos_y, other.pos_x)

        def render_tile(self, screen: pygame.Surface) -> None:
            """
            Blit the tile image to the provided pygame screen.

            Args:
                screen (pygame.Surface): Surface on which the tile is drawn.
            """
            screen.blit(self.image, (self.pos_x * self.width, self.pos_y * self.height))

        def get_pos(self) -> Tuple[int, int, int]:
            """
            Get the (x, y, level) position of the tile.
            
            Returns:
                Tuple[int, int, int]: The position of the tile.
            """
            return (self.pos_x, self.pos_y, self.level)

    # -------------------------------------------------------------------------
    class Button:
        """
        Implements a clickable button for the problem's UI.
        
        Attributes:
            rect (pygame.Rect): The rectangular region occupied by the button.
            text (str): The text displayed on the button.
            callback (Callable): Function to invoke when the button is activated.
            color (tuple): Default background color.
            hover_color (tuple): Background color when the mouse hovers.
            font (pygame.font.Font): Font used to render button text.
            surface (pygame.Surface): Surface that supports alpha transparency.
        """
        def __init__(self, x: int, y: int, w: int, h: int, text: str, callback) -> None:
            self.rect = pygame.Rect(x, y, w, h)
            self.text = text
            self.callback = callback
            self.color = (0, 0, 0, 150)
            self.hover_color = (0, 100, 0)
            self.font = pygame.font.SysFont("Comic Sans MS", 24)
            self.surface = pygame.Surface((w, h), pygame.SRCALPHA)

        def draw(self, screen: pygame.Surface) -> None:
            """
            Draw the button (with hover effect) onto the given screen.
            
            Args:
                screen (pygame.Surface): The surface to render the button on.
            """
            mouse_pos = pygame.mouse.get_pos()
            is_hovered = self.rect.collidepoint(mouse_pos)
            self.surface.fill(self.hover_color if is_hovered else self.color)
            screen.blit(self.surface, self.rect.topleft)
            text_surface = self.font.render(self.text, True, (255, 255, 255))
            screen.blit(text_surface, text_surface.get_rect(center=self.rect.center))

        def handle_event(self, event: pygame.event.Event) -> None:
            """
            Process mouse and keyboard events for the button.
            
            Args:
                event (pygame.event.Event): The event to handle.
            """
            if event.type == pygame.MOUSEBUTTONDOWN and self.rect.collidepoint(event.pos):
                self.callback()
            elif event.type == pygame.KEYDOWN and event.key == pygame.K_SPACE:
                self.callback()

    # -------------------------------------------------------------------------
    class Level:
        """
        Represents a single level (floor) of the map.
        
        Attributes:
            num (int): The level index.
            visible (bool): Flag indicating whether the level is currently rendered.
            grid (List[List[Optional[problem.Tile]]]): 2D array of Tile objects.
            width (int): Number of tiles in the horizontal direction.
            height (int): Number of tiles in the vertical direction.
            texts (List[Tuple[pygame.Surface, Tuple[int, int]]]): Overlay text or icons.
            source (Optional[Tuple[int, int, int]]): The selected starting tile.
            destination (Optional[Tuple[int, int, int]]): The selected target tile.
        """
        def __init__(self, num: int, data: List[List[Optional["Problem.Tile"]]], width: int, height: int,
                     texts: List[Tuple[pygame.Surface, Tuple[int, int]]]) -> None:
            self.num = num
            self.visible = False
            self.grid = data
            self.width = width
            self.height = height
            self.texts = texts
            self.source: Optional[Tuple[int, int, int]] = None
            self.destination: Optional[Tuple[int, int, int]] = None
            self.fix_stairs()  # Adjust stair tile elevations

        def get_tile(self, x: int, y: int) -> Optional["Problem.Tile"]:
            """
            Retrieve a tile from the grid at coordinates (x, y) if within bounds.
            
            Args:
                x (int): x-coordinate index.
                y (int): y-coordinate index.
            
            Returns:
                Optional[problem.Tile]: The tile at (x, y) or None if out of bounds.
            """
            if 0 <= x < self.width and 0 <= y < self.height:
                return self.grid[x][y]
            return None

        def toggle_visibility(self) -> None:
            """
            Toggle the visibility flag of the level.
            """
            self.visible = not self.visible

        def set_visibility(self, flag: bool) -> None:
            """
            Set the visibility flag of the level.
            
            Args:
                flag (bool): True for visible, False otherwise.
            """
            self.visible = flag

        def fix_stairs(self) -> None:
            """
            Adjust the elevation (z_level) of tiles designated as stairs.
            
            Two groups of stairs (A and B) are processed separately to assign consistent
            elevation values for cross-level navigation.
            """
            stairs_a_indices = [(46, 29, 0), (46, 28, 0), (46, 27, 0), (46, 26, 0), (46, 25, 0),
                                  (47, 25, 0), (48, 25, 0), (48, 26, 0), (48, 27, 0), (48, 28, 0)]
            stairs_b_indices = [(5, 35, 0), (4, 35, 0), (3, 35, 0), (2, 35, 0), (1, 35, 0),
                                  (1, 34, 0), (1, 33, 0), (2, 33, 0), (3, 33, 0), (4, 33, 0)]
            inc_a = 1 / (len(stairs_a_indices) + 1)
            for i, pos in enumerate(stairs_a_indices):
                tile = self.grid[pos[0]][pos[1]]
                if tile:
                    tile.z_level = round((i + 1) * inc_a, 2)
            inc_b = 1 / len(stairs_b_indices)
            for i, (x, y, _) in enumerate(stairs_b_indices):
                tile = self.grid[x][y]
                if tile:
                    tile.z_level = round((i + 1) * inc_b, 2)

        def render_level(self, screen: pygame.Surface, agents: Optional[List["Problem.Agent"]]) -> None:
            """
            Render the level including its tiles, overlay texts, and markers for
            the source and destination. Also draws the trails of any agents currently on this level.
            
            Args:
                screen (pygame.Surface): The screen to render on.
                agents (Optional[List[problem.Agent]]): A list of agent objects to draw trails.
            """
            # Draw a semi-transparent overlay to darken the background.
            overlay = pygame.Surface(screen.get_size(), pygame.SRCALPHA)
            overlay.fill((0, 0, 0, 100))
            screen.blit(overlay, (0, 0))
            # Render each tile.
            for y in range(self.height):
                for x in range(self.width):
                    tile = self.grid[x][y]
                    if tile:
                        tile.render_tile(screen)
            # Render overlay texts.
            for surface, pos in self.texts:
                screen.blit(surface, pos)
            # Render each agent's trail if the agent is on this level.
            if agents:
                for agent in agents:
                    if agent.level == self.num:
                        agent.draw_trail()
            # Render the source and destination markers if set.
            if self.source is not None and self.source[2] == self.num:
                self.render_source(screen, self.source)
            if self.destination is not None and self.destination[2] == self.num:
                self.render_destination(screen, self.destination)
            

        def render_source(self, screen: pygame.Surface, pos: Tuple[int, int, int]) -> None:
            """
            Render a circle marker for the source (starting tile) on the level.
            
            Args:
                screen (pygame.Surface): The surface to render on.
                pos (Tuple[int, int, int]): The (x, y, level) position of the source tile.
            """
            # Create a confirmation button (recreated every frame here; consider caching it).
            self.confirm_button = Problem.Button(100, 100, 200, 40, "Confirm", Problem.confirm_selection)
            self.source = pos
            x, y, _ = pos
            tile = self.get_tile(x, y)
            highlight_color = (255, 255, 0)  # Yellow marker.
            if tile and self.visible:
                center_x, center_y = Problem.get_tile_center(tile)
                pygame.draw.circle(screen, (0,0,0), (center_x, center_y), self.height // 8)
                pygame.draw.circle(screen, highlight_color, (center_x, center_y), self.height // 10)

        def render_destination(self, screen: pygame.Surface, pos: Tuple[int, int, int]) -> None:
            """
            Render a circle marker for the destination tile on the level.
            
            Args:
                screen (pygame.Surface): The surface to render on.
                pos (Tuple[int, int, int]): The (x, y, level) position of the destination tile.
            """
            self.confirm_button = Problem.Button(100, 100, 200, 40, "Confirm", Problem.confirm_selection)
            self.destination = pos
            x, y, _ = pos
            tile = self.get_tile(x, y)
            highlight_color = (0, 120, 200)  # Blue-ish marker.
            if tile and self.visible:
                center_x, center_y = Problem.get_tile_center(tile)
                pygame.draw.circle(screen, (0,0,0), (center_x, center_y), self.height // 8)
                pygame.draw.circle(screen, highlight_color, (center_x, center_y), self.height // 10)

    # -------------------------------------------------------------------------
    class Map:
        """
        Represents the full multi-level map loaded from a TMX file.

        Attributes:
            problem (problem): A reference to the main problem object.
            data: The pytmx loaded map data.
            width (int): Number of tiles horizontally.
            height (int): Number of tiles vertically.
            levels (Dict[int, problem.Level]): Dictionary mapping level index to Level object.
            level_texts (Dict[int, List[Tuple[pygame.Surface, Tuple[int, int]]]]): Overlays for each level.
        """
        def __init__(self, problem: "Problem", map_path: str) -> None:
            self.problem = problem
            self.data = load_pygame(map_path)
            self.width = self.data.width
            self.height = self.data.height
            self.levels: Dict[int, Problem.Level] = {}
            self.level_texts: Dict[int, List[Tuple[pygame.Surface, Tuple[int, int]]]] = {0: [], 1: []}

            # Set up overlay texts.
            font = pygame.font.SysFont("Comic Sans MS", 24)
            emoji_font = pygame.font.SysFont("Segoe UI Emoji", 24)
            self.level_texts[0] = [
                (font.render("Class001", True, (0, 0, 0)), (34, 80)),
                (font.render("Class001", False, (0, 0, 0)), (34, 80)),
                (font.render("Parking", False, (0, 0, 0)), (587, 57)),
                (font.render("Class002", False, (0, 0, 0)), (34, 690)),
                (font.render("Class003", False, (0, 0, 0)), (357, 560)),
                (font.render("Class004", False, (0, 0, 0)), (355, 702)),
                (font.render("Class005", False, (0, 0, 0)), (663, 690)),
                (emoji_font.render("â™¿", True, (0, 0, 0)), (34, 380)),
                (font.render("Cafe", False, (0, 0, 0)), (660, 380)),
                (emoji_font.render("ðŸ›—", True, (0, 0, 0)), (431, 468)),
                (emoji_font.render("â¬†ï¸", True, (0, 0, 0)), (40, 540)),
                (emoji_font.render("â¬†ï¸", True, (0, 0, 0)), (746, 420))
            ]
            self.level_texts[1] = [
                (font.render("Library", False, (0, 0, 0)), (365, 660)),
                (emoji_font.render("â™¿", True, (0, 0, 0)), (390, 540)),
                (emoji_font.render("â™¿", True, (0, 0, 0)), (540, 540)),
                (emoji_font.render("ðŸ›—", True, (0, 0, 0)), (431, 468)),
                (emoji_font.render("â¬‡ï¸", True, (0, 0, 0)), (40, 540)),
                (emoji_font.render("â¬‡ï¸", True, (0, 0, 0)), (746, 420)),
                (font.render("Student Center", False, (0, 0, 0)), (320, 390)),
                (font.render("Class101", False, (0, 0, 0)), (64, 310)),
                (font.render("Class102", False, (0, 0, 0)), (600, 310))
            ]

            # Build the level grid based on visible tile layers.
            for l in range(sum(1 for i in self.data.visible_tile_layers)):
                grid: List[List[Optional[Problem.Tile]]] = [[None for _ in range(self.width)] for _ in range(self.height)]
                for y in range(self.height):
                    for x in range(self.width):
                        if self.data.get_tile_gid(x, y, l):
                            grid[x][y] = Problem.Tile(
                                x, y, l,
                                self.data.get_tile_image(x, y, l),
                                self.data.get_tile_properties(x, y, l)  # type: ignore
                            )
                        else:
                            grid[x][y] = None
                self.levels[l] = Problem.Level(l, grid, self.width, self.height, self.level_texts[l])

        def get_tile(self, x: int, y: int, level: int) -> Optional["Problem.Tile"]:
            """
            Retrieve the tile at the specified grid position and level.
            
            Args:
                x (int): x-coordinate.
                y (int): y-coordinate.
                level (int): Level index.
            
            Returns:
                Optional[problem.Tile]: The tile if found, else None.
            """
            return self.levels[level].grid[x][y]

        def render_map(self, screen: pygame.Surface, agents: Optional[List["Problem.Agent"]] = None) -> None:
            """
            Render all visible levels along with agent trails (if provided).
            
            Args:
                screen (pygame.Surface): The screen surface.
                agents (Optional[List[problem.Agent]]): List of agents to draw.
            """
            for lvl in self.levels.values():
                if lvl.visible:
                    lvl.render_level(screen, agents)

        def render_source(self, screen: pygame.Surface, pos: Tuple[int, int, int]) -> None:
            """
            Render the source marker on every level.
            """
            for lvl in self.levels.values():
                lvl.render_source(screen, pos)

        def render_destination(self, screen: pygame.Surface, pos: Tuple[int, int, int]) -> None:
            """
            Render the destination marker on every level.
            """
            for lvl in self.levels.values():
                lvl.render_destination(screen, pos)

        def render_press_start_screen(self, screen: pygame.Surface) -> None:
            """
            Render the initial screen prompting the user to press Enter.
            """
            screen.fill((30, 30, 30))
            try:
                font = pygame.font.SysFont('Arial', 48)
            except Exception:
                pygame.font.init()
                font = pygame.font.Font(None, 48)
            text = font.render("Press Enter to Start", True, (255, 255, 255))
            text_rect = text.get_rect(center=(screen.get_width() // 2, screen.get_height() // 2))
            screen.blit(text, text_rect)

        def toggle_level1(self) -> None:
            """
            Toggle the visibility of level 1 and update the rendered map.
            """
            self.levels[1].toggle_visibility()
            print("Toggle Visibility for Level1")
            self.render_map(self.problem.screen, [self.problem.a_agent, self.problem.d_agent])
            self.problem.a_agent.draw_trail()
            self.problem.d_agent.draw_trail()

    # -------------------------------------------------------------------------
    class PriorityQueue:
        """
        Implements a priority queue using Python's heapq module.

        Items are stored as tuples (priority, item) and popped in sorted order.
        """
        def __init__(self, order: str = 'min', f=lambda x: x) -> None:
            self.heap: List[Tuple[Any, Any]] = []
            if order == 'min':
                self.f = f
            elif order == 'max':
                self.f = lambda x: -f(x)
            else:
                raise ValueError("Order must be either 'min' or 'max'.")

        def append(self, item: Any) -> None:
            heapq.heappush(self.heap, (self.f(item), item))

        def extend(self, items: List[Any]) -> None:
            for item in items:
                self.append(item)

        def pop(self) -> Any:
            if self.heap:
                return heapq.heappop(self.heap)[1]
            else:
                raise Exception('Trying to pop from empty PriorityQueue.')

        def __len__(self) -> int:
            return len(self.heap)

        def __contains__(self, key: Any) -> bool:
            return any(item == key for _, item in self.heap)

        def __getitem__(self, key: Any) -> Any:
            for value, item in self.heap:
                if item == key:
                    return value
            raise KeyError(f"{key} is not in the priority queue")

        def __delitem__(self, key: Any) -> None:
            try:
                index = [item == key for _, item in self.heap].index(True)
                del self.heap[index]
            except ValueError:
                raise KeyError(f"{key} is not in the priority queue")
            heapq.heapify(self.heap)

    # -------------------------------------------------------------------------
    class StateType(Enum):
        """
        Enumerates the various states of the problem.
        """
        PressStart = auto()
        SelectStarting = auto()
        SelectDestination = auto()
        Analyze = auto()
        ShowResult = auto()
        VisualizeResult = auto()

    class State:
        """
        Manages and holds the current problem state.
        """
        def __init__(self) -> None:
            self.current_state: Problem.StateType = Problem.StateType.PressStart

        @property
        def state(self) -> "Problem.StateType":
            return self.current_state

        @state.setter
        def state(self, new_state: "Problem.StateType") -> None:
            print(f"ðŸ” Switching state: {self.current_state.name} â†’ {new_state.name}")
            self.current_state = new_state

        def __eq__(self, state_type: "Problem.StateType") -> bool:
            return self.current_state == state_type

    # -------------------------------------------------------------------------
    class Agent:
        """
        Represents an agent that navigates the map along a computed path.

        Attributes:
            x, y (int): Current grid coordinates.
            level (int): Current level.
            path (List[Tuple[int,int,int]]): The computed path as (x, y, level) tuples.
            trail (List[Tuple[int,int,int]]): History of positions for drawing the path trail.
            color (tuple): Color for rendering the agent.
            size (int): Marker size for the agent.
            delay (int): Delay (in ms) between path updates (for visualization).
        """
        def __init__(self, problem: "Problem", x: int = 25, y: int = 0, level: int = 0,
                     color: Tuple[int, int, int] = (0, 0, 255)) -> None:
            self.x = x
            self.y = y
            self.level = level
            self.path: List[Tuple[int, int, int]] = []
            self.color = color
            self.problem = problem
            self.map = self.problem.map
            self.trail: List[Tuple[int, int, int]] = []
            self.size = 4
            self.path_index = 0
            self.delay = int((1 / self.problem.get_agent_speed()) * 100)

        def set_path(self, path: List[Tuple[int, int, int]]) -> None:
            """
            Set the agent's path and reset its trail.

            Args:
                path (List[Tuple[int, int, int]]): The new path as a list of (x, y, level).
            """
            self.path = path
            self.path_index = 0
            self.trail = []

        def update(self) -> None:
            """
            Advance the agent one step along its path.
            """
            if self.path_index < len(self.path):
                self.x, self.y, self.level = self.path[self.path_index]
                self.trail.append((self.x, self.y, self.level))
                self.path_index += 1

        def draw_trail(self) -> None:
            """
            Render the historical trail of the agent on the screen.
            """
            for x, y, level in self.trail:
                tile = self.map.get_tile(x, y, level)
                if tile:
                    visible_level = 1 if self.map.levels[1].visible else 0
                    if level == visible_level:
                        screen_x, screen_y = Problem.get_tile_center(tile)
                        pygame.draw.circle(self.problem.screen, self.color, (screen_x, screen_y), tile.width // self.size)

        def draw(self) -> None:
            """
            Draw the agent's current position.
            """
            tile = self.map.get_tile(self.x, self.y, self.level)
            if tile:
                screen_x, screen_y = Problem.get_tile_center(tile)
                pygame.draw.circle(self.problem.screen, self.color, (screen_x, screen_y), tile.width // self.size)

        def draw_path(self) -> None:
            """
            Animate the agent moving along its path.
            """
            while not self.is_done():
                self.update()
                self.draw_trail()
                self.draw()
                pygame.display.flip()
                pygame.time.delay(self.delay)

        def is_done(self) -> bool:
            """
            Check if the agent has finished following the path.

            Returns:
                bool: True if the end is reached, else False.
            """
            return self.path_index >= len(self.path)

    # -------------------------------------------------------------------------
    class Analyzer:
        """
        Runs the A* and Dijkstra pathfinding algorithms on the map.
        
        Attributes:
            map (problem.Map): The map on which to compute paths.
            start (problem.Tile): The starting tile.
            end (problem.Tile): The destination tile.
            results (Dict[str, Any]): Stores computed paths for each algorithm.
            top_results (Dict[str, Any]): Stores the best result for each algorithm type.
        """
        def __init__(self, map_obj: "Problem.Map", start: "Problem.Tile", end: "Problem.Tile") -> None:
            self.map = map_obj
            self.start = start
            self.end = end
            self.results: Dict[str, Any] = {}
            self.top_results: Dict[str, Any] = {}

        def is_same_level(self) -> bool:
            """
            Check if the start and end tiles reside on the same level.

            Returns:
                bool: True if both are on the same level.
            """
            return self.start.level == self.end.level

        import time  # Make sure to import time at the top of your file

        def run_algorithms(self) -> None:
            """
            Execute A* (with various heuristics) and Dijkstra algorithms.
            Record the execution time and path cost for each run.
            Then, choose the best overall path based on:
                - BestOverall: comparing the product of cost and runtime.
                - BestCost: comparing cost primarily, then runtime.
                - BestTime: comparing runtime primarily, then cost.
            Store each result in self.results under different keys.
            """
            heuristics = {
                "Manhattan": self.heuristic_manhattan,
                "Euclidean": self.heuristic_euclidean,
                "Chebyshev": self.heuristic_chebyshev,
            }

            # Variables to hold best A* run metrics.
            best_astar_path = None
            best_astar_cost = float('inf')
            best_astar_runtime = float('inf')
            best_astar_method = None

            # Run A* using each heuristic and record both path cost and runtime.
            for name, heuristic in heuristics.items():
                start_time = time.perf_counter()
                path = self.run_astar(heuristic)
                end_time = time.perf_counter()
                runtime = end_time - start_time
                if path:
                    cost = len(path)
                    self.results[f"A*_{name}"] = (path, cost, runtime)
                    # Compare lex order: (cost, runtime)
                    if (cost, runtime) < (best_astar_cost, best_astar_runtime):
                        best_astar_cost = cost
                        best_astar_runtime = runtime
                        best_astar_path = path
                        best_astar_method = f"A*_{name}"

            self.top_results["A*"] = (best_astar_method, best_astar_cost, best_astar_runtime, best_astar_path)

            # Run Dijkstra
            best_dk_path = None
            best_dk_cost = float('inf')
            best_dk_runtime = float('inf')
            start_time = time.perf_counter()
            dijkstra_path = self.run_dijkstra()
            end_time = time.perf_counter()
            dk_runtime = end_time - start_time
            if dijkstra_path:
                cost = len(dijkstra_path)
                self.results["Dijkstra"] = (dijkstra_path, cost, dk_runtime)
                best_dk_cost = cost
                best_dk_runtime = dk_runtime
                best_dk_path = dijkstra_path

            self.top_results["Dijkstra"] = (best_dk_cost, best_dk_runtime, best_dk_path)

            # Only proceed if both algorithms produced a valid path.
            if best_astar_path is not None and best_dk_path is not None:
                # BestOverall: Compare the product (cost * runtime).
                overall_astar = best_astar_cost * best_astar_runtime
                overall_dk = best_dk_cost * best_dk_runtime
                if overall_astar < overall_dk:
                    best_overall_method = best_astar_method
                    best_overall_cost = best_astar_cost
                    best_overall_runtime = best_astar_runtime
                    best_overall_path = best_astar_path
                elif overall_astar > overall_dk:
                    best_overall_method = "Dijkstra"
                    best_overall_cost = best_dk_cost
                    best_overall_runtime = best_dk_runtime
                    best_overall_path = best_dk_path
                else:
                    best_overall_method = "Same"
                    best_overall_cost = best_astar_cost
                    best_overall_runtime = best_astar_runtime
                    best_overall_path = best_astar_path or best_dk_path

                # BestCost: Compare (cost, runtime).
                if (best_astar_cost, best_astar_runtime) < (best_dk_cost, best_dk_runtime):
                    best_cost_method = best_astar_method
                    best_cost_cost = best_astar_cost
                    best_cost_runtime = best_astar_runtime
                    best_cost_path = best_astar_path
                elif (best_astar_cost, best_astar_runtime) > (best_dk_cost, best_dk_runtime):
                    best_cost_method = "Dijkstra"
                    best_cost_cost = best_dk_cost
                    best_cost_runtime = best_dk_runtime
                    best_cost_path = best_dk_path
                else:
                    best_cost_method = "Same"
                    best_cost_cost = best_astar_cost
                    best_cost_runtime = best_astar_runtime
                    best_cost_path = best_astar_path or best_dk_path

                # BestTime: Compare (runtime, cost).
                if (best_astar_runtime, best_astar_cost) < (best_dk_runtime, best_dk_cost):
                    best_time_method = best_astar_method
                    best_time_cost = best_astar_cost
                    best_time_runtime = best_astar_runtime
                    best_time_path = best_astar_path
                elif (best_astar_runtime, best_astar_cost) > (best_dk_runtime, best_dk_cost):
                    best_time_method = "Dijkstra"
                    best_time_cost = best_dk_cost
                    best_time_runtime = best_dk_runtime
                    best_time_path = best_dk_path
                else:
                    best_time_method = "Same"
                    best_time_cost = best_astar_cost
                    best_time_runtime = best_astar_runtime
                    best_time_path = best_astar_path or best_dk_path

                # Print all three results.
                print(f"âœ… BestOverall: {best_overall_method} with cost {best_overall_cost} and composite score {overall_astar:.5f} (if A*: {overall_astar:.5f} vs Dijkstra: {overall_dk:.5f})")
                print(f"âœ… BestCost: {best_cost_method} with cost {best_cost_cost} and runtime {best_cost_runtime:.5f} seconds")
                print(f"âœ… BestTime: {best_time_method} with runtime {best_time_runtime:.5f} seconds and cost {best_time_cost}")

                # Store the three results.
                self.top_results["BestOverall"] = (best_overall_method, best_overall_cost, best_overall_runtime, overall_astar)
                self.top_results["BestCost"] = (best_cost_method, best_cost_path, best_cost_cost, best_cost_runtime)
                self.top_results["BestTime"] = (best_time_method, best_time_path, best_time_runtime, best_time_cost)
            else:
                print("âŒ Could not find a valid path.")

                
        def get_result(self, algorithm_name: str) -> Any:
            """
            Get the path result for a specific algorithm.
            
            Args:
                algorithm_name (str): The algorithm's name.
                
            Returns:
                Any: The associated result (path) if exists, else None.
            """
            return self.results.get(algorithm_name)

        def get_neighbors(self, tile: "Problem.Tile", map_obj: "Problem.Map", current_level: int, use_terrain_cost: bool = True) -> List[Tuple["Problem.Tile", float]]:
            """
            Retrieve neighboring tiles for a given tile, including cross-level neighbors for
            special zones (e.g., elevators and stairs).

            Args:
                tile (problem.Tile): The current tile.
                map_obj (problem.Map): The map object.
                current_level (int): The current level of the tile.
                use_terrain_cost (bool): Whether to factor terrain costs.

            Returns:
                List[Tuple[problem.Tile, float]]: A list of (neighbor, cost) pairs.
            """
            x, y = tile.pos_x, tile.pos_y
            z = tile.z_level
            zone = tile.zone.lower()
            width, height = map_obj.width, map_obj.height
            current_grid = map_obj.levels[current_level].grid
            neighbors: List[Tuple[Problem.Tile, float]] = []

            # Check same-level neighbors (4 directions).
            for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
                nx, ny = x + dx, y + dy
                if 0 <= nx < width and 0 <= ny < height:
                    neighbor = current_grid[nx][ny]
                    if neighbor and neighbor.zone.lower() != "red":
                        elevation_diff = abs(neighbor.z_level - z)
                        if elevation_diff <= 0.15:
                            cost = Problem.calculate_cost(tile, neighbor, use_terrain_cost)
                            neighbors.append((neighbor, cost))
            # Handle cross-level neighbors for elevators (purple) or stairs (yellow).
            other_level = 1 if current_level == 0 else 0
            if zone == "purple":
                n_tile = map_obj.get_tile(x, y, other_level)
                if n_tile:
                    neighbors.append((n_tile, 0))
            elif zone == "yellow":
                for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
                    nx, ny = x + dx, y + dy
                    if 0 <= nx < width and 0 <= ny < height:
                        neighbor = map_obj.get_tile(nx, ny, other_level)
                        if neighbor and neighbor.zone.lower() != "red":
                            elevation_diff = abs(neighbor.z_level - z)
                            if elevation_diff <= 0.15:
                                cost = Problem.calculate_cost(tile, neighbor, use_terrain_cost)
                                neighbors.append((neighbor, cost))
            return neighbors

        def run_dijkstra(self) -> Optional[List[Tuple[int, int, int]]]:
            """
            Run Dijkstra's algorithm to compute the shortest path.

            Returns:
                Optional[List[Tuple[int, int, int]]]: The path as a list of (x, y, level) tuples, or None if no path is found.
            """
            start = self.start
            goal = self.end
            frontier = Problem.PriorityQueue(order='min', f=lambda x: x[1])
            frontier.append((start, 0))
            came_from: Dict[Problem.Tile, Optional[Problem.Tile]] = {start: None}
            cost_so_far: Dict[Problem.Tile, float] = {start: 0}
            while len(frontier) > 0:
                current, _ = frontier.pop()
                if current == goal:
                    break
                for neighbor, move_cost in self.get_neighbors(current, self.map, current.level,
                                                                self.map.problem.get_use_friction_for_dijkstra()):
                    new_cost = cost_so_far[current] + move_cost
                    if neighbor not in cost_so_far or new_cost < cost_so_far[neighbor]:
                        cost_so_far[neighbor] = new_cost
                        frontier.append((neighbor, new_cost))
                        came_from[neighbor] = current
            path: List[Tuple[int, int, int]] = []
            tile = goal
            while tile != start:
                path.append((tile.pos_x, tile.pos_y, tile.level))
                tile = came_from.get(tile)
                if tile is None:
                    return None
            path.append((start.pos_x, start.pos_y, start.level))
            path.reverse()
            return path

        def heuristic_manhattan(self, a: "Problem.Tile", b: "Problem.Tile") -> float:
            """
            Compute the Manhattan distance between two tiles.
            """
            dx = abs(a.pos_x - b.pos_x)
            dy = abs(a.pos_y - b.pos_y)
            dz = abs(a.level - b.level)
            return dx + dy + dz

        def heuristic_euclidean(self, a: "Problem.Tile", b: "Problem.Tile") -> float:
            """
            Compute the Euclidean distance between two tiles.
            """
            dx = a.pos_x - b.pos_x
            dy = a.pos_y - b.pos_y
            dz = a.z_level - b.z_level
            return math.sqrt(dx * dx + dy * dy + dz * dz)

        def heuristic_chebyshev(self, a: "Problem.Tile", b: "Problem.Tile") -> float:
            """
            Compute the Chebyshev distance between two tiles.
            """
            dx = abs(a.pos_x - b.pos_x)
            dy = abs(a.pos_y - b.pos_y)
            dz = abs(a.z_level - b.z_level)
            return max(dx, dy, dz)

        def run_astar(self, heuristic_func) -> Optional[List[Tuple[int, int, int]]]:
            """
            Run the A* algorithm with the specified heuristic function to compute the shortest path.

            Args:
                heuristic_func (Callable): A function that calculates a heuristic distance between two tiles.

            Returns:
                Optional[List[Tuple[int, int, int]]]: The computed path as (x, y, level) tuples or None if no path exists.
            """
            start = self.start
            goal = self.end
            frontier = Problem.PriorityQueue(order='min', f=lambda x: x[1])
            frontier.append((start, 0))
            came_from: Dict[Problem.Tile, Optional[Problem.Tile]] = {start: None}
            cost_so_far: Dict[Problem.Tile, float] = {start: 0}
            while len(frontier) > 0:
                current, _ = frontier.pop()
                if current == goal:
                    break
                for neighbor, move_cost in self.get_neighbors(current, self.map, current.level, use_terrain_cost=True):
                    new_cost = cost_so_far[current] + move_cost
                    if neighbor not in cost_so_far or new_cost < cost_so_far[neighbor]:
                        cost_so_far[neighbor] = new_cost
                        priority = new_cost + self.map.problem.get_heuristic_multiplier() * heuristic_func(neighbor, goal)
                        frontier.append((neighbor, priority))
                        came_from[neighbor] = current
            path: List[Tuple[int, int, int]] = []
            tile = goal
            while tile != start:
                path.append((tile.pos_x, tile.pos_y, tile.level))
                tile = came_from.get(tile)
                if tile is None:
                    return None
            path.append((start.pos_x, start.pos_y, start.level))
            path.reverse()
            return path

    # -------------------------------------------------------------------------
    # End of Inner Classes
    # -------------------------------------------------------------------------

    # -------------------------------------------------------------------------
    # problem Instance Attributes and Methods
    # -------------------------------------------------------------------------
    def __init__(self, map_path = "Tiled/Map.tmx" ,gui=True) -> None:
        # Private configuration parameters.
        self.__use_friction_for_dijkstra = True
        self.__agent_speed = 2         # Agent speed: allowed range 1 to 10.
        self.__heuristic_multiplier = 2.0  # Multiplier for weighting the A* heuristic.
        
        self.is_gui = gui

        # Initialize problem state and pygame components.
        self.state = Problem.State()
        pygame.init() 
        self.screen = pygame.display.set_mode((800, 800))
        pygame.display.set_caption("Wheel Chair Map Navigation")
        self.clock = pygame.time.Clock()
        self.source: Tuple[Optional[int], Optional[int], Optional[int]] = (None, None, None)
        self.destination: Tuple[Optional[int], Optional[int], Optional[int]] = (None, None, None)

        # Create the map instance.
        self.map = Problem.Map(self, map_path)
        self.set_reset_parameters()

    # Public setter and getter methods for configuration.
    def set_use_friction_for_dijkstra(self, value: bool) -> None:
        self.__use_friction_for_dijkstra = value

    def get_use_friction_for_dijkstra(self) -> bool:
        return self.__use_friction_for_dijkstra

    def set_agent_speed(self, speed: int) -> None:
        self.__agent_speed = speed

    def get_agent_speed(self) -> int:
        return self.__agent_speed

    def set_heuristic_multiplier(self, multiplier: float = 2.0) -> None:
        self.__heuristic_multiplier = multiplier

    def get_heuristic_multiplier(self) -> float:
        return self.__heuristic_multiplier

    def get_results(self) -> Dict[str, Any]:
        """
        Public getter method to retrieve the results from the Analyzer.
        
        Returns:
            Dict[str, Any]: The results from the analyzer if computed, else an empty dictionary.
        """
        
        if hasattr(self, "analyzer"):
             
            out =  {result:
                (
                    str(self.analyzer.results[result][1])+" steps" ,
                 str(round(self.analyzer.results[result][2] *1000,4)) + "ms"
                 ) 
                for result  in self.analyzer.results}
            return out
      
        else:
            return {}
        
    def get_top_results(self) -> Dict[str, Any]:
        if hasattr(self, "analyzer"):   
            return self.analyzer.top_results
        else: 
            return {}
    def set_reset_parameters(self) -> None:
        """
        Initialize or reset the problem parameters and objects at the start of the problem.
        """
        self.running = True
        self.y_offset = 100
        self.toggle_button = Problem.Button(0, 0, 200, 40, "Toggle Level 1", self.map.toggle_level1)
        self.visualize_button = Problem.Button(280, 700, 200, 40, "Visualize", self.visualize)
        self.source_selected = False
        self.source_confirmed = False
        self.destination_selected = False
        self.destination_confirmed = False
        self.analysis_result_text = ""
        self.a_agent = Problem.Agent(self)
        self.d_agent = Problem.Agent(self)
        # Create a default Analyzer with arbitrary initial source and destination.
        self.analyzer = Problem.Analyzer(self.map, self.map.get_tile(25, 0, 0), self.map.get_tile(25, 0, 0))  # type: ignore

    def run(self) -> None:
        """
        The main problem loop which handles events, state transitions, and rendering.
        """
        if not self.is_gui:
            pygame.quit()
            if not self.source: print("Source Unspecified")
            if not self.destination: print("Destination Unspecified")
            if self.source and self.destination:
                self.analyze()
            return 
        while self.running and self.is_gui:
            for event in pygame.event.get():
                # Handle toggle events in selection states.
                if self.state in [Problem.StateType.SelectStarting, Problem.StateType.SelectDestination]:
                    self.toggle_button.handle_event(event)
                # Handle UI events during result display.
                if self.state in [Problem.StateType.ShowResult, Problem.StateType.VisualizeResult]:
                    self.toggle_button.handle_event(event)
                    self.visualize_button.handle_event(event)
                if event.type == pygame.QUIT:
                    self.running = False
                    pygame.quit()
                elif event.type == pygame.MOUSEBUTTONDOWN:
                    self.handle_mouse_click(pygame.mouse.get_pos())
                elif event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_RETURN:
                        self.confirm_selection()
                    if event.key == pygame.K_ESCAPE and self.state == Problem.StateType.VisualizeResult:
                        self.state = Problem.StateType.PressStart
                    if event.key == pygame.K_RETURN and self.state == Problem.StateType.VisualizeResult:
                        self.state = Problem.StateType.ShowResult

            # Render content based on the current problem state.
            match self.state:
                case Problem.StateType.PressStart:
                    self.set_reset_parameters()
                    self.map.render_press_start_screen(self.screen)
                    keys = pygame.key.get_pressed()
                    if keys[pygame.K_RETURN]:
                        self.state = Problem.StateType.SelectStarting

                case Problem.StateType.SelectStarting:
                    self.map.levels[0].set_visibility(True)
                    self.select_source()

                case Problem.StateType.SelectDestination:
                    self.select_destination()

                case Problem.StateType.Analyze:
                    self.render_analyze_screen()
                    pygame.display.flip()
                    pygame.time.delay(500)
                    _ = self.analyze()
                    self.state = Problem.StateType.ShowResult

                case Problem.StateType.ShowResult:
                    self.show_results()

                case Problem.StateType.VisualizeResult:
                    self.toggle_button.draw(self.screen)
                    self.visualize_button.draw(self.screen)

            pygame.display.flip()
            self.clock.tick(60)
        pygame.quit()

    def show_results(self) -> None:
        """
        Render an overlay showing the results of the pathfinding analysis.
        """
        overlay = pygame.Surface(self.screen.get_size(), pygame.SRCALPHA)
        self.map.render_map(self.screen)
        overlay.fill((0, 0, 0, 150))
        self.screen.blit(overlay, (0, 0))
        font = pygame.font.SysFont("Arial", 22)
        self.y_offset = 100
        a_star_data = self.analyzer.top_results["A*"]
        dijkstra_data = self.analyzer.top_results["Dijkstra"]
        results_text = [
            f"A* Heuristic: {a_star_data[0]}  Path Cost: {a_star_data[1]} with time taken: {a_star_data[2]*1000}ms",
            f"Dijkstra Path Cost: {dijkstra_data[0]} with time taken: {dijkstra_data[1]*1000}ms",
            "Press Visualize to See paths"
        ]
        print(results_text[:-1])
        for line in results_text:
            rendered = font.render(line, True, (255, 255, 255))
            self.screen.blit(rendered, (80, self.y_offset))
            self.y_offset += 80
        self.visualize_button.draw(self.screen)
        self.state = Problem.StateType.VisualizeResult

    def visualize(self) -> None:
        """
        Animate the agents moving along the computed paths for A* and Dijkstra.
        """
        a_path = self.analyzer.top_results['A*'][3]
        d_path = self.analyzer.top_results['Dijkstra'][2]
        print("Visualizing")
        print("A*: blue")
        print("Dijkstra: Yellow")
        self.a_agent.set_path(a_path)
        self.d_agent.set_path(d_path)
        self.a_agent.size = 3
        self.d_agent.size = 4
        self.d_agent.color = (255, 255, 0)
        self.screen.fill((0, 0, 0))
        self.map.render_map(self.screen)
        self.toggle_button.draw(self.screen)
        self.a_agent.draw_path()
        self.d_agent.draw_path()

    def render_analyze_screen(self) -> None:
        """
        Render an overlay with text to indicate that pathfinding is in progress.
        """
        self.map.render_map(self.screen)
        font = pygame.font.SysFont("Arial", 30)
        sub_font = pygame.font.SysFont("Arial", 24)
        pygame.draw.rect(self.screen, (0, 0, 0, 180), pygame.Rect(150, 300, 500, 200))
        analyzing_text = font.render("Analyzing Path...", True, (255, 255, 255))
        self.screen.blit(analyzing_text, analyzing_text.get_rect(center=(400, 340)))
        loading_text = sub_font.render("Please wait while algorithms evaluate", True, (180, 180, 180))
        self.screen.blit(loading_text, loading_text.get_rect(center=(400, 390)))

    def analyze(self) -> Optional[Any]:
        """
        Run the pathfinding analysis using the Analyzer object and return the best path.

        Returns:
            Optional[Any]: The best path if found, otherwise None.
        """
        print("ðŸ” Analyzing path...")
        if not self.source or not self.destination:
            print("âŒ Invalid start or end tile.")
            self.analysis_result_text = "âŒ Invalid start or end tile."
            return None
        start_tile = self.map.get_tile(self.source[0], self.source[1], self.source[2])  # type: ignore
        end_tile = self.map.get_tile(self.destination[0], self.destination[1], self.destination[2])  # type: ignore
        self.analyzer = Problem.Analyzer(self.map, start_tile, end_tile)  # type: ignore
        self.analyzer.run_algorithms()
        result = self.analyzer.get_result("Best")
        return result

    def select_source(self) -> None:
        """
        Render the selection screen for choosing the source tile and display a hint.
        """
        self.map.render_map(self.screen)
        self.toggle_button.draw(self.screen)
        if self.source_selected:
            font = pygame.font.SysFont("Arial", 24)
            hint_text = font.render("Press ENTER to confirm source", True, (255, 255, 255))
            self.screen.blit(hint_text, (225, 750))

    def select_destination(self) -> None:
        """
        Render the selection screen for choosing the destination tile and display a hint.
        """
        self.map.render_map(self.screen)
        self.toggle_button.draw(self.screen)
        if self.destination_selected:
            font = pygame.font.SysFont("Arial", 24)
            hint_text = font.render("Press ENTER to confirm Destination", True, (255, 255, 255))
            self.screen.blit(hint_text, (225, 750))

    def confirm_selection(self) -> None:
        """
        Confirm the source or destination selection and update the problem state accordingly.
        """
        if self.state == Problem.StateType.SelectStarting and self.source_selected:
            self.source_confirmed = True
            print("Source Confirmed")
            self.state = Problem.StateType.SelectDestination
        if self.state == Problem.StateType.SelectDestination and self.destination_selected:
            self.destination_confirmed = True
            print("Destination Confirmed")
            self.state = Problem.StateType.Analyze

    def handle_mouse_click(self, pos: Tuple[int, int]) -> None:
        """
        Process a mouse click to set the source or destination tile based on current state.

        Args:
            pos (Tuple[int, int]): The (x, y) coordinates of the mouse click on screen.
        """
        tile_x = pos[0] // self.map.data.tilewidth
        tile_y = pos[1] // self.map.data.tileheight
        if self.state == Problem.StateType.SelectStarting:
            level = 1 if self.map.levels[1].visible else 0
            self.source = (tile_x, tile_y, level)
            self.source_selected = True
            self.map.render_source(self.screen, self.source)
            print(f"ðŸ“ Source selected at {self.source}")
        elif self.state == Problem.StateType.SelectDestination:
            level = 1 if self.map.levels[1].visible else 0
            self.destination = (tile_x, tile_y, level)
            self.destination_selected = True
            self.map.render_destination(self.screen, self.destination)
            print(f"ðŸ“ Destination selected at {self.destination}")

```

```python
import matplotlib.pyplot as plt
import random
import pandas as pd
```

```python
# Define a range of heuristic multipliers to test, for example from 1 to 5 in increments of 0.5.
multiplier_range = [x * 0.5 for x in range(2, 11)]  # 1.0, 1.5, ..., 5.0
overall_product = []  # to store a combined metric for each multiplier
problem_2 = Problem("Tiled/Map.tmx")
pygame.quit()
for mult in multiplier_range:
    
    problem_2.set_heuristic_multiplier(mult)
    # You might want to configure start and destination, e.g.:
    problem_2.source = source
    problem_2.destination = destination
    # Run the analysis algorithm (this will update problem_2.analyzer and top_results)
    problem_2.analyze()
    # Get, for instance, the overall composite score stored in BestOverall (cost * runtime)
    if "BestOverall" in problem_2.analyzer.top_results:
        # Unpack: (method, path, cost, runtime, composite)
        _, cost, runtime, composite = problem_2.get_top_results()["BestOverall"]
        
        overall_product.append(composite)
    else:
        overall_product.append(None)

# Create a plot of heuristic multiplier vs composite metric.
plt.figure(figsize=(8, 5))
plt.plot(multiplier_range, overall_product, marker='o')
plt.xlabel("Heuristic Multiplier")
plt.ylabel("Composite Metric (cost * runtime)")
plt.title("Optimal Heuristic Multiplier for Best Overall Path")
plt.grid(True)
plt.show()
best_index = overall_product.index(min(overall_product))
best_multiplier = multiplier_range[best_index]
print(f"Best heuristic multiplier for optimal path: {best_multiplier}")

```

```python
import random
problem_3 = Problem("Tiled/Map.tmx")
pygame.quit()

# Function to pick a random (x,y,z)
def random_coordinate() -> Tuple[int, int, int]:
    while True:
        level = random.randint(0, 1)
        if level == 0:
            coord = (random.randint(0, 49), random.randint(0, 49), level)
        else:
            coord = (random.randint(0, 49), random.randint(16, 49), level)
        tile = problem_3.map.get_tile(coord[0], coord[1], coord[2])
        if tile is not None and tile.zone.lower() != "red":
                return coord

num_trials = 100
results_list = []

for run_idx in range(num_trials):
    problem_3.source,problem_3.destination = None,None # type: ignore
    # Randomly choose coordiantes
    src_coord = random_coordinate()
    dst_coord = random_coordinate()
    # Assemble full coordinates as (x, y, z)
    while problem_3.source== None and problem_3.destination == None:
        problem_3.source = (src_coord[0], src_coord[1], src_coord[2])
        problem_3.destination = (dst_coord[0], dst_coord[1], dst_coord[2])
    
    # Run the analysis (which should update problem_3.analyzer.results)
    problem_3.analyze()
    
    # For each algorithm variant (A* variants and Dijkstra) we expect that analyze() stores results in
    # problem_3.analyzer.results as dictionaries with keys like "A*_Manhattan", etc.
    # Each value is assumed to be a tuple: (path, cost, runtime).
    # We then calculate composite = cost * runtime.
    # For each algorithm variant, try to get the result tuple: (path, cost, runtime)
    # and calculate the composite metric = cost * runtime.
    try:
        manhattan_result = problem_3.analyzer.results["A*_Manhattan"]
        manhattan_cost = manhattan_result[1]
        manhattan_runtime = manhattan_result[2]
        manhattan_composite = manhattan_cost * manhattan_runtime
    except KeyError:
        manhattan_composite, manhattan_cost, manhattan_runtime = None, None, None

    try:
        euclidean_result = problem_3.analyzer.results["A*_Euclidean"]
        euclidean_cost = euclidean_result[1]
        euclidean_runtime = euclidean_result[2]
        euclidean_composite = euclidean_cost * euclidean_runtime
    except KeyError:
        euclidean_composite, euclidean_cost, euclidean_runtime = None, None, None

    try:
        chebyshev_result = problem_3.analyzer.results["A*_Chebyshev"]
        chebyshev_cost = chebyshev_result[1]
        chebyshev_runtime = chebyshev_result[2]
        chebyshev_composite = chebyshev_cost * chebyshev_runtime
    except KeyError:
        chebyshev_composite, chebyshev_cost, chebyshev_runtime = None, None, None

    try:
        dijkstra_result = problem_3.analyzer.results["Dijkstra"]
        dijkstra_cost = dijkstra_result[1]
        dijkstra_runtime = dijkstra_result[2]
        dijkstra_composite = dijkstra_cost * dijkstra_runtime
    except KeyError:
        dijkstra_composite, dijkstra_cost, dijkstra_runtime = None, None, None

    # Store all values in the results list.
    results_list.append({
        "Run": run_idx,
        "Source": problem_3.source,
        "Destination": problem_3.destination,
        "A*_Manhattan": manhattan_composite,
        "A*_Manhattan_cost": manhattan_cost,
        "A*_Manhattan_time": manhattan_runtime,
        "A*_Euclidean": euclidean_composite,
        "A*_Euclidean_cost": euclidean_cost,
        "A*_Euclidean_time": euclidean_runtime,
        "A*_Chebyshev": chebyshev_composite,
        "A*_Chebyshev_cost": chebyshev_cost,
        "A*_Chebyshev_time": chebyshev_runtime,
        "Dijkstra": dijkstra_composite,
        "Dijkstra_cost": dijkstra_cost,
        "Dijkstra_time": dijkstra_runtime
    })

    
    # For a single-instance run, do not quit pygame each time.
    # You can re-run problem_3.analyze() for each trial since problem_3 is already initialized.

```

```python
df = pd.DataFrame(results_list)

# Plot a separate line graph for each algorithm's composite metric versus run index.
window_size = 5  # Adjust as needed
df["A*_Manhattan_smooth"] = df["A*_Manhattan"].rolling(window=window_size, center=True).mean()
df["A*_Euclidean_smooth"] = df["A*_Euclidean"].rolling(window=window_size, center=True).mean()
df["A*_Chebyshev_smooth"] = df["A*_Chebyshev"].rolling(window=window_size, center=True).mean()
df["Dijkstra_smooth"] = df["Dijkstra"].rolling(window=window_size, center=True).mean()

# Now plot these smoothed columns
plt.figure(figsize=(10, 6))
plt.plot(df["Run"], df["A*_Manhattan_smooth"], label="A* Manhattan (smoothed)")
plt.plot(df["Run"], df["A*_Euclidean_smooth"], label="A* Euclidean (smoothed)")
plt.plot(df["Run"], df["A*_Chebyshev_smooth"], label="A* Chebyshev (smoothed)")
plt.plot(df["Run"], df["Dijkstra_smooth"], label="Dijkstra (smoothed)")

plt.xlabel("Run Index")
plt.ylabel("Composite Metric (Cost * Runtime) [Smoothed]")
plt.title("Smoothed Composite Metric vs. Run Index")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Initialize win counters.
wins = {"A*_Manhattan": 0, "A*_Euclidean": 0, "A*_Chebyshev": 0, "Dijkstra": 0}

# Loop through each row in the DataFrame.
for idx, row in df.iterrows():
    # Collect composite values if they are not None.
    comps = {}
    if pd.notna(row["A*_Manhattan"]):
        comps["A*_Manhattan"] = row["A*_Manhattan"]
    if pd.notna(row["A*_Euclidean"]):
        comps["A*_Euclidean"] = row["A*_Euclidean"]
    if pd.notna(row["A*_Chebyshev"]):
        comps["A*_Chebyshev"] = row["A*_Chebyshev"]
    if pd.notna(row["Dijkstra"]):
        comps["Dijkstra"] = row["Dijkstra"]
    
    if comps:
        # Get the algorithm with the lowest composite value.
        best_algo = min(comps, key=comps.get) # type: ignore
        wins[best_algo] += 1

# Create a bar graph from the win data.
algorithms = list(wins.keys())
win_counts = list(wins.values())
plt.figure(figsize=(8, 5))
bars = plt.bar(algorithms, win_counts, color=["blue", "green", "orange", "red"])
plt.xlabel("Algorithm")
plt.ylabel("Win Count (Lowest Composite)")
plt.title("Number of Wins per Algorithm over Trials")
plt.grid(axis="y")
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{int(height)}', 
             ha='center', va='bottom', fontsize=12, color='black')
plt.show()

```

```python
import numpy as np
from typing import List, Union
# Create subplots for each algorithm.
fig, axes = plt.subplots(2, 2, figsize=(10, 8), sharex=True, sharey=True)

# Define a helper function to plot regression line.
def plot_regression(ax, x_data, y_data, color="black"):
    # Remove NaN values (if any) so regression doesn't error out.
    mask = ~np.isnan(x_data) & ~np.isnan(y_data)
    x = x_data[mask]
    y = y_data[mask]
    if len(x) > 0:
        # Compute linear regression coefficients: slope and intercept
        coeffs = np.polyfit(x, y, 1)
        slope, intercept = coeffs
        # Generate fitted y values.
        x_fit = np.linspace(x.min(), x.max(), 100)
        y_fit = slope * x_fit + intercept
        ax.plot(x_fit, y_fit, color=color, linestyle="-", label="Fit")
        # Optionally, you could add text with the slope/intercept:
        ax.text(0.05, 0.95, f"m={slope:.2f}", transform=ax.transAxes,
                fontsize=9, verticalalignment='top')

# Plot for A* Manhattan.
x_manhattan = df["A*_Manhattan_time"].values
y_manhattan = df["A*_Manhattan_cost"].values
axes[0,0].scatter(x_manhattan, y_manhattan, color="blue")
plot_regression(axes[0,0], x_manhattan, y_manhattan,color="blue")
axes[0,0].set_title("A* Manhattan")

# Plot for A* Euclidean.
x_euclidean = df["A*_Euclidean_time"].values
y_euclidean = df["A*_Euclidean_cost"].values
axes[0,1].scatter(x_euclidean, y_euclidean, color="orange")
plot_regression(axes[0,1], x_euclidean, y_euclidean, color="orange")
axes[0,1].set_title("A* Euclidean")

# Plot for A* Chebyshev.
x_chebyshev = df["A*_Chebyshev_time"].values
y_chebyshev = df["A*_Chebyshev_cost"].values
axes[1,0].scatter(x_chebyshev, y_chebyshev, color="green")
plot_regression(axes[1,0], x_chebyshev, y_chebyshev, color="green")
axes[1,0].set_title("A* Chebyshev")

# Plot for Dijkstra.
x_dijkstra = df["Dijkstra_time"].values
y_dijkstra = df["Dijkstra_cost"].values
axes[1,1].scatter(x_dijkstra, y_dijkstra, color="red")
plot_regression(axes[1,1], x_dijkstra, y_dijkstra, color="red")
axes[1,1].set_title("Dijkstra")

# Set labels for each subplot.
for ax in axes.flat:
    ax.set(xlabel="Runtime (s)", ylabel="Path Cost")

plt.tight_layout()
plt.show()
```
